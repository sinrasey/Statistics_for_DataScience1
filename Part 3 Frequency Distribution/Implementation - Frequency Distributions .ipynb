### Exploring the Data

The data set is about basketball players in WNBA (Women's National Basketball Association), and contains general information about players, along with their metrics for the season 2016-2017.

```python
wnba = pd.read_csv('DataSet/wnba.csv')
```

```python
wnba.shape
```

```python
wnba.describe()
```

```python
sns.boxplot(wnba['Height'])
plt.yticks(np.arange(160, 210, 5))
plt.show()
```

```python
Height_wnba = wnba['Height'].value_counts().sort_index()
```

```python
100*Height_wnba/len(wnba)
```

```python
len(Height_wnba.index)
```

```python
206-165
```

```python
Height_wnba.values
```

```python
plt.figure(figsize=(12, 6))
plt.xlabel('x')
plt.ylabel('Frequency')
plt.plot(Height_wnba.index, Height_wnba.values, color = 'b')
plt.title('Frequency of Height')
plt.xticks(np.arange(165, 210, step=10))
plt.yticks(np.arange(0, 22, step=2))
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(Height_wnba.index, Height_wnba.values, color = 'r')
plt.title('Height Frequency')
plt.xticks(np.arange(165, 210, step=3))
plt.yticks(np.arange(1, 22, step=2))
plt.grid()
for i, value in enumerate(Height_wnba.values):
    plt.text(Height_wnba.index[i], value, str(value), ha='center', va='bottom')
plt.show()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.barh(Height_wnba.index, Height_wnba.values, color = 'r')
plt.title('Height Frequency')
plt.xticks(np.arange(0, 22, step=2))
plt.yticks(np.arange(165, 210, step=3))
plt.grid()
plt.show()
```

### Histogram:

**Type of Data:** When is a histogram most suitable, and for what kind of data?

**Purpose:** What does a histogram communicate about the distribution of values in a dataset?

**Appropriate Data:** How do you determine whether a histogram is the right choice for visualizing your data?

```python
wnba['BMI'].value_counts()
```

```python
sns.histplot(wnba['BMI'], bins=20, color='teal', kde=True)
plt.xticks(np.arange(18, 32, 1))
plt.yticks(np.arange(0, 26, 2))
plt.show()
```

```python
plt.hist(wnba['BMI'],bins=15, edgecolor = 'white', rwidth= 0.97, color='r')
plt.show()
```

```python

# Create a figure and axis
fig, ax = plt.subplots()

# Create a histogram for each row in the data
# ax.hist(data, bins=7, edgecolor='white',
#         rwidth=0.999, color = 'teal' )
ax.hist(wnba['BMI'], edgecolor = 'white', 
       rwidth= 0.99, color='b')
ax.set_xlabel('Value with class frequency')
ax.set_ylabel('Frequency in range')
ax.set_xticks(np.arange(18, 32, 1))
ax.set_title('Frequency Distribution')

plt.show()
```

### Area Plot:

**Type of Data:** In what scenarios is an area plot a suitable choice, and what kind of data does it represent?

**Purpose:** How does an area plot illustrate the cumulative values or proportions of different variables over time or categories?

**Appropriate Data:** When might an area plot be more informative than other types of visualizations for temporal or categorical data?

```python
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]

# Basic stacked area chart.
plt.stackplot(x,y1, y2, y3, labels=['A','B','C'])
plt.legend(loc='upper left')
```
